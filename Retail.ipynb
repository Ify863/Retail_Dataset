## EXPLORATORY DATA ANALYSIS

### 1.What is the total sales revenue over time (daily, weekly, monthly)?

```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Ensure 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index (useful for resampling)
df.set_index('Date', inplace=True)

# ðŸ”¹ Daily Sales
daily_sales = df['Total Amount'].resample('D').sum()

# ðŸ”¹ Weekly Sales
weekly_sales = df['Total Amount'].resample('W').sum()

# ðŸ”¹ Monthly Sales
monthly_sales = df['Total Amount'].resample('M').sum()

# Plotting
plt.figure(figsize=(14, 8))

# Daily
plt.subplot(3, 1, 1)
sns.lineplot(x=daily_sales.index, y=daily_sales.values, color='teal')
plt.title('Daily Sales Revenue')
plt.ylabel('Total Sales')
plt.xlabel('Date')

# Weekly
plt.subplot(3, 1, 2)
sns.lineplot(x=weekly_sales.index, y=weekly_sales.values, color='navy')
plt.title('Weekly Sales Revenue')
plt.ylabel('Total Sales')
plt.xlabel('Date')

# Monthly
plt.subplot(3, 1, 3)
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, color='darkgreen')
plt.title('Monthly Sales Revenue')
plt.ylabel('Total Sales')
plt.xlabel('Date')

plt.tight_layout()
plt.show()
```

### 2.which product category generates the most revenue?


```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Group by Product Category and sum the Total Amount
category_sales = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.values, y=category_sales.index, palette='viridis')

plt.title('Total Revenue by Product Category')
plt.xlabel('Total Sales Revenue')
plt.ylabel('Product Category')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Print top-earning category
top_category = category_sales.idxmax()
top_revenue = category_sales.max()
print(f"âœ… Highest revenue generated by: '{top_category}' â†’ â‚¦{top_revenue:,.2f}")

```

### 3.What is the gender distribution of sales?

```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Group sales by gender
gender_sales = df.groupby('Gender')['Total Amount'].sum()

# Print the result
print("ðŸ’¡ Gender-wise Sales Revenue:")
print(gender_sales)

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', colors=['skyblue', 'lightpink'], startangle=90)
plt.title('Sales Distribution by Gender (Pie Chart)')
plt.axis('equal')  # Keeps the pie circular
plt.show()

# Plot bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_sales.index, y=gender_sales.values, palette='pastel')
plt.title('Sales Revenue by Gender (Bar Chart)')
plt.ylabel('Total Sales')
plt.xlabel('Gender')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 4.Which age group spends the most?


```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# ðŸ“Š Create age groups
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['<18', '18â€“25', '26â€“35', '36â€“45', '46â€“60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# ðŸ”¢ Total sales per age group
age_group_sales = df.groupby('Age Group')['Total Amount'].sum().sort_values(ascending=False)

# ðŸ–¼ï¸ Plot bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=age_group_sales.index, y=age_group_sales.values, palette='coolwarm')

plt.title('Total Sales Revenue by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# ðŸ¥‡ Top spending age group
top_group = age_group_sales.idxmax()
top_value = age_group_sales.max()
print(f"âœ… Age group that spends the most: '{top_group}' with â‚¦{top_value:,.2f} in sales.")
```

### 5.Who are the top 10 highest spending customers

```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# ðŸ”¢ Group by customer and sum their total spend
top_customers = df.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)

# ðŸ–¼ï¸ Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='mako')

plt.title('Top 10 Highest Spending Customers')
plt.xlabel('Total Sales Amount')
plt.ylabel('Customer ID')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ðŸ¥‡ Print top customer info
top_customer = top_customers.idxmax()
top_value = top_customers.max()
print(f"âœ… Top spending customer: '{top_customer}' with â‚¦{top_value:,.2f} in purchases.")
```

### 6.Which products have the highest quantity sold?


```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Group by Product Category and sum Quantity sold
quantity_by_product = df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=quantity_by_product.values, y=quantity_by_product.index, palette='crest')

plt.title('Quantity Sold by Product Category')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Category')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Print top product category
top_product = quantity_by_product.idxmax()
top_quantity = quantity_by_product.max()
print(f"âœ… Product category with highest quantity sold: '{top_product}' with {top_quantity} units.")
```

### 7. What is the relationship between price and quantity sold?


```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Price per Unit', y='Quantity', hue='Product Category', palette='tab10', alpha=0.7)

plt.title('Relationship Between Price and Quantity Sold')
plt.xlabel('Price per Unit')
plt.ylabel('Quantity Sold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 8.Are there seasonalÂ trends?

```python


# Load your dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Ensure 'Date' column is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a 'Month' column (e.g., Jan, Feb...)
df['Month'] = df['Date'].dt.month_name()

# Optional: Sort months chronologically
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Group by month and sum sales
monthly_sales = df.groupby('Month')['Total Amount'].sum().reindex(month_order)

# Plot seasonal trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='steelblue')

plt.title('Monthly Sales Trend (Seasonal Pattern)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 10.  Which day of the week has the highest sales volume?

```python


# Load dataset
df = pd.read_csv(r"C:\Users\DELL\Desktop\PYTHON\retail_sales_dataset.csv")

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the weekday name (e.g., Monday, Tuesday...)
df['Weekday'] = df['Date'].dt.day_name()

# Group by weekday and sum sales
weekday_sales = df.groupby('Weekday')['Total Amount'].sum()

# Sort the days in week order
weekday_sales = weekday_sales.reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=weekday_sales.index, y=weekday_sales.values, palette='Set2')

plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# Optional: print the top-performing day
top_day = weekday_sales.idxmax()
top_sales = weekday_sales.max()
print(f"âœ… The day with the highest sales is **{top_day}**, totaling â‚¦{top_sales:,.2f}.")
```
